```python
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS

embedding_model = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

vectorstore = FAISS.from_documents(splits, embedding_model)
print("Vector store created successfully!")
```

```python
retriever = vectorstore.as_retriever(
    search_type="similarity",
    search_kwargs={"k": 4}
)
```

```python
from langchain_core.prompts import ChatPromptTemplate

rag_prompt = ChatPromptTemplate.from_template("""
You are a legal document assistant.
Answer the question strictly using the provided context.
Cite the relevant section numbers when possible.
If the answer is not explicitly stated, say: "Not found in the document."

Context:
{context}

Question:
{question}

Answer:
""")

```

```python
from langchain_google_genai import ChatGoogleGenerativeAI

llm = ChatGoogleGenerativeAI(
    model="gemini-3-flash-preview",
    temperature=0.2
)
```

```python
def format_docs(docs):
    return "\n\n".join(
        f"(Page {doc.metadata.get('page', 'N/A')})\n{doc.page_content}"
        for doc in docs
    )
```

```python
from langchain_core.runnables import RunnablePassthrough
from langchain_core.output_parsers import StrOutputParser

rag_chain = (
    {
        "context": retriever | format_docs,
        "question": RunnablePassthrough()
    }
    | rag_prompt
    | llm
    | StrOutputParser()
)
```

```python
rag_chain.invoke("What is the employee salary?")
```

```python
summary_prompt = ChatPromptTemplate.from_template("""
You are a legal assistant.

Provide a professional summary of the following employment contract.
Highlight:
- Parties involved
- Compensation
- Termination terms
- Confidentiality clauses
- Key obligations

Contract:
{context}

Summary:
""")

summary_chain = summary_prompt | llm | StrOutputParser()

full_text = format_docs(splits)

contract_summary = summary_chain.invoke({"context": full_text})

print(contract_summary)
```

```python
test_questions = [
    "What is the employee salary?",
    "What are the termination conditions?",
    "Is there a confidentiality clause?",
    "What are the employee obligations?"
]

for q in test_questions:
    print("Question:", q)
    print("Answer:", rag_chain.invoke(q))
    print("=" * 70)
```

```python
sample_question = "What is the employee salary?"

retrieved_docs = retriever.invoke(sample_question)

for i, doc in enumerate(retrieved_docs):
    print(f"Chunk {i+1}")
    print("Page:", doc.metadata.get("page"))
    print(doc.page_content[:300])
    print("-" * 50)
```

```python
import gradio as gr

def chat_fn(question):
    return rag_chain.invoke(question)

demo = gr.Interface(
    fn=chat_fn,
    inputs=gr.Textbox(label="Ask about the document"),
    outputs=gr.Textbox(label="Answer"),
    title="Smart Contract RAG Assistant"
)

demo.launch(share=True)
```
